```python
def test8(a):
    l = []
    for i in a:
        if type(i) == list:
            for j in i:
                l.append(j)
        else :
            if type(i) == int or type(i) == float :
                l.append(i)
    return l 
```

```python

test8(l)
```

```python
def test9(a):
    """this is my fuction to extract num data from list"""
    l = []
    for i in a:
        if type(i) == list:
            for j in i:
                l.append(j)
        else :
            if type(i) == int or type(i) == float :
                l.append(i)
    return l 
```
