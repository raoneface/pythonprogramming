```python
l = [1,2,3,34,4,[2,3,4] , "sudh" , "kumar"]
```

```python
l1_int = []
l2_str = []

for i in l :
    logging.info("we are iterating through our list and our local var is i")
    if type(i)== list :
        logging.info("i am inside if statement and i am trying to check list type")
        for j in i :
            logging.info("i am another for loop for list inside list element")
            if type(j)  ==  int :
                logging.info("i am inside if statement")
                l1_int.append(j)
    elif type(i) == int :
        l1_int.append(i)
    else :
        if type(i) == str :
            l2_str.append(i)
```
