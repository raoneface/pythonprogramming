```python
import io
```

```python
with open("test1.txt" ,"wb") as f :
    file = io.BufferedWriter(f)

    file.write(b"this is my first line\n ")
    file.write(b"this is my second line\n ")
    file.flush()
```

```python
with open("test1.txt","rb") as f :
    file = io.BufferedReader(f)
    data = file.read()
```
