```python
import numpy as np
import pandas as pd
import seaborn as sns 
```

```python

df = sns.load_dataset("healthexp")
df.head()
```

```python
## covariance
df = df.select_dtypes(include=np.number)
df.cov()
```

```python
## pearson correlation
df.corr(method="pearson")
```

```python
## rank correlation
df.corr(method="spearman")
```
