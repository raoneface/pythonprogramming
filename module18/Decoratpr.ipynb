# decorator

```python
def decofun(func):
    def inner_dec():
        print("this is start my function")
        func()
        print("This is end of function")
    return inner_dec
```

```python
@decofun
def test1():
    print(7+14)
```

```python
test1()
```

```python
import time
def decorator1(func):
    def timer_inner_deco():
        start_time = time.time()
        func()
        end_time = time.time()
        print(end_time-start_time)
    return timer_inner_deco
```

```python
@decorator1
def test2():
    print(45+36)
```

```python
test2()
```

```python
@decorator1
def test3():
    for i in range(100000000):
        pass
```

```python
test3()
```
